```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(1, figsize=(19,13))

# Sample it down to only the Java Island region
lon_min, lon_max = 105.21, 114.60527777778
lat_min, lat_max = -8.78, -5.8758333333333

# Mercator of Java Island
m1 = Basemap(projection='merc',
             llcrnrlat=lat_min,
             urcrnrlat=lat_max,
             llcrnrlon=lon_min,
             urcrnrlon=lon_max)

m1.drawmapboundary(fill_color='#A6CAE0', linewidth=0)
m1.fillcontinents(color='grey', alpha=0.7, lake_color='grey')
m1.drawcoastlines(linewidth=0.1, color="white")

lons = [112.922038, 112.949927, 112.589468, 111.194039, 110.440814, 110.445826, 110.072347, 109.995808, 109.932328,
       109.220557, 108.406708, 107.861161, 107.837799, 107.730874, 108.072516, 106.982081, 106.964891, 106.733284,
       106.051871, 110.890938, 112.574908, 112.450009, 112.313951, 111.760129, 113.566461, 114.046669, 114.242246,
       112.620118]
lats = [-8.107205, -7.941377, -7.763777, -7.626906, -7.452502, -7.540684, -7.384580, -7.301866, -7.196916, -7.236928,
       -6.893102, -7.320253, -7.149210, -7.315101, -7.263552, -6.787113, -6.768102, -6.706401, -6.257500, -6.616094,
       -7.731838, -7.919443, -7.928837, -7.813484, -7.963123, -8.119561, -8.056086, -7.613921]
labels = ['MT. Semeru', 'MT. Bromo', 'MT. Arjuno', 'MT. Lawu', 'MT. Merbabu', 'MT. Merapi', 'MT. Sumbing', 'MT. Sindoro',
        'MT. Prau', 'MT. Selamet', 'MT. Ciremai', 'MT. Cikuray', 'MT. Guntur', 'MT. Papandayan', 'MT. Galunggung', 
        'MT. Gede', 'MT. Pangrango', 'MT. Salak', 'MT. Karang', 'MT. Muria', 'MT. Welirang', 'MT. Butak', 'MT. Kelud',
        'MT. Wilis', 'MT. Argopuro', 'MT. Raung', 'MT. Ijen', 'MT. Penanggungan']

x_offsets = [10000, -30000, -25000]
y_offsets = [5000, -25000, -25000]

x,y = m1(lons, lats)
m1.plot(x, y, 'bo', markersize=10)

plt.title("Pemetaan Gunung di Pulau Jawa")

for label, x, y in zip(labels, x, y):
    plt.text(x+10000, y+5000, label)
    
plt.show()
```
